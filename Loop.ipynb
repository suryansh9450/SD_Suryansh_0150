```python
l = [1,2,3,4,4,"suryansh","sharad",254,23.5]

```

```python
l1_num =[]
l2_str = []
for i in l:
    if type(i) == int or type(i) == float:
        l1_num.append(i)
    else:
        l2_str.append(i)
```
